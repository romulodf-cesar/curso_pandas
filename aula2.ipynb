# Representação da Média

```python
# importe o pandas 
import pandas as pd
```

```python
# carregue o dataset_ruido.csv
df = pd.read_csv('datasets/dataset_ruido.csv')
```

```python
df['nivel ruido (dB)'].mean()
```

```python
# renomeie a coluna nivel ruido (dB) para nivel_ruido_db
df = df.rename(columns={'nivel ruido (dB)': 'nivel_ruido_db'})
```

```python
df.nivel_ruido_db.mean()
```

```python
# ordene e calcule a mediana agora
df = df.sort_values(by='nivel_ruido_db')
df.nivel_ruido_db.median()
```
